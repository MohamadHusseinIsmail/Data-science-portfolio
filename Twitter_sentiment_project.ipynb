```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
Social_data= pd.read_csv("/content/twitter_training.csv")
Social_data.head()
```

```python
Social_data.columns = ['ID', 'Topic', 'Sentiment', 'Text']
```

```python
sentiment_distribution = Social_data['Sentiment'].value_counts()


print(sentiment_distribution)
```

```python
# Plot the sentiment distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_distribution.index, y=sentiment_distribution.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
filtered_data = Social_data[Social_data['Topic'] == 'Borderlands']


filtered_sentiment_distribution = filtered_data['Sentiment'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=filtered_sentiment_distribution.index, y=filtered_sentiment_distribution.values, palette='viridis')
plt.title('Sentiment Distribution for Borderlands')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
