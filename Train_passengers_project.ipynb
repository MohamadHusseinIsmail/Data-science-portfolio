```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv('/content/train( Prodgy infotech).csv')
print(df.head(10))
```

```python
print(df.info())


```

```python
print(df.isnull().sum())

```

```python

print(df.describe())


```

```python
print(df.describe(include='object'))
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Distribution of numerical variables
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Ages')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=df)
plt.title('Count of Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Pclass', data=df)
plt.title('Scatter Plot between Numeric Column 1 and Numeric Column 2')
plt.show()

# Calculate correlation
correlation = df[['Age', 'Pclass']].corr()
print(correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Survived', data=df)
plt.title('Box Plot of Survived passengers by Gender')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Survived', 'Pclass']])
plt.show()
```
